```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df=pd.read_csv("polygon.csv")

df.head()
```

```python
# Count missing values before cleaning
missing_values_before_cleaning = df.isnull().sum()

# Display the count of missing values
missing_values_before_cleaning
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
filtered_date_hist = df[df["DateTime_(UTC)"] >= "2024-04-16"]
df.loc[filtered_date_hist.index, "Historical_$Price/MATIC"] = filtered_date_hist["CurrentValue_@_$0.703172221747878/MATIC"] / filtered_date_hist["Value_IN(MATIC)"]
df = df.dropna(subset=["Historical_$Price/MATIC"])
```

```python
fig = px.line(df, 
              x="DateTime_(UTC)", 
              y="Historical_$Price/MATIC",
              title="Line plot of Historical $Price/MATIC",
              labels={"DateTime_(UTC)": "Date", "Historical_$Price/MATIC": "$Price/MATIC"})
fig.show()
```

```python
# Filters for each month
first_month_filter = df[df["DateTime_(UTC)"] < "2024-03-01"]
second_month_filter = df[(df["DateTime_(UTC)"] >= "2024-03-01") & (df["DateTime_(UTC)"] < "2024-04-01") ]
third_month_filter = df[df["DateTime_(UTC)"] >= "2024-04-01"]

# Statistics
# Mean

first_month_historical_mean = first_month_filter["Historical_$Price/MATIC"].mean()
second_month_historical_mean = second_month_filter["Historical_$Price/MATIC"].mean()
third_month_historical_mean = third_month_filter["Historical_$Price/MATIC"].mean()

# Mean amount

first_month_amount_mean = first_month_filter["Value_IN(MATIC)"].mean()
second_month_amount_mean = second_month_filter["Value_IN(MATIC)"].mean()
third_month_amount_mean = third_month_filter["Value_IN(MATIC)"].mean()

# Standard Deviation
first_month_historical_std = first_month_filter["Historical_$Price/MATIC"].std()
second_month_historical_std = second_month_filter["Historical_$Price/MATIC"].std()
third_month_historical_std = third_month_filter["Historical_$Price/MATIC"].std()



print("Statistics of MATIC from 2024-02-08 to 2024-04-16")
print(f"\nMean Price in February: {first_month_historical_mean:.2f} $/MATIC  "  
      f"Standard Deviation: {first_month_historical_std:.2f} $/MATIC  " 
      f"Average MATIC bought: {first_month_amount_mean:.2f}" )
print(f"Mean Price in March: {second_month_historical_mean:.2f} $/MATIC  "  
      f"Standard Deviation: {second_month_historical_std:.2f} $/MATIC  "
      f"Average MATIC bought: {second_month_amount_mean:.2f}")
print(f"Mean Price in April: {third_month_historical_mean:.2f} $/MATIC  "  
      f"Standard Deviation: {third_month_historical_std:.2f} $/MATIC  "
      f"Average MATIC bought: {third_month_amount_mean:.2f}")
```

```python
# Convert 'DateTime (UTC)' to date format and extract the day of the week
df.loc[:, 'DayOfWeek'] = pd.to_datetime(df['Corrected_Column_Name']).dt.day_name()

# Count the number of transactions per day of the week
day_counts = df['DayOfWeek'].value_counts().reset_index()
day_counts.columns = ['DayOfWeek', 'Count']

# Calculate the percentage for each day
day_counts['Percentage'] = (day_counts['Count'] / day_counts['Count'].sum()) * 100

# Create a pie chart with hover effect
fig = px.pie(day_counts, values='Count', names='DayOfWeek', title='Distribution of Transactions by Day of the Week',
             hover_data={'DayOfWeek': True, 'Count': True, 'Percentage': ':.1f'})

# Customize hover template
fig.update_traces(hovertemplate='<b>%{label}</b><br>Count: %{value}<br>Percentage: %{customdata[2]:.1f}%')

# Show the plot
fig.show()
```
